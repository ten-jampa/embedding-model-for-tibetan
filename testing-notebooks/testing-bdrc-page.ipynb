```python
import requests
```

```python
url = "https://www.lotsawahouse.org/bo/tibetan-masters/tsongkhapa/praise-to-sarasvati"
```

```python
set_of_urls = {
    'praise-to-sarasvati': 'https://www.lotsawahouse.org/bo/tibetan-masters/tsongkhapa/praise-to-sarasvati',
    'foundation-of-all-qualities': "https://www.lotsawahouse.org/bo/tibetan-masters/tsongkhapa/foundation-of-all-qualities",
    'gesar-windhorse-invocation': "https://www.lotsawahouse.org/bo/tibetan-masters/mipham/brief-gesar-windhorse-invocation"
}
```

```python
key_to_text = {}
for key, url in set_of_urls.items():
    resp = requests.get(url)
    text = resp.text
    fixed = text.encode('latin-1').decode('utf-8')
    key_to_text[key] = fixed
```

```python
print(key_to_text['gesar-windhorse-invocation'])
```
